```python
import pandas as pd
import gensim
import warnings
warnings.simplefilter('ignore')
```

# データの読み込み

```python
df = pd.read_csv('preprocessed_lyrics.csv')
df.head()
```

# 学習

```python
def remove_zero_len_word(lyric):
    return [word for word in lyric if len(word)]
```

```python
lyrics = list(df['remove_stop_words'])
lyrics = [liryc.split(' ') for liryc in lyrics]
lyrics = [remove_zero_len_word(liryc) for liryc in lyrics]
lyrics[0]
```

```python
model = gensim.models.word2vec.Word2Vec(
    lyrics,
    size=100,  # ベクトルの次元
    min_count=3,  # 最小3回以上出ている単語のみの使用
    window=5,  # window幅
    iter=4000,  # 学習の繰り返し数
    seed=2018)
```

```python
len(model.wv.vocab), model.wv.vocab
```

```python
model.save('bz2vec.model')
```

# 類似度の高い単語

```python
def get_similor_words(model, word):
    results = model.wv.most_similar(positive=word, topn=10)
    df_result = pd.DataFrame(results, columns=['Word', 'Sim'])
    return df_result
```

```python
get_similor_words(model, '愛')
```

```python
get_similor_words(model, '恋')
```

# 単語の足し算 / 引き算

```python
def get_positive_negative_word(positive, negative):
    results = model.most_similar(positive=positive, negative=negative, topn=10)
    df_result = pd.DataFrame(results, columns=['Word', 'Sim'])
    return df_result
```

```python
get_positive_negative_word(positive=['恋', '未来'], negative=['あなた'])
```
